```python
import sys
sys.path.insert(0, '../code/')
import pickle
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
%matplotlib inline
```

```python
res = pickle.load(open('../scripts/results_mnist/exp_2_R_05.p','rb'))
res_conf = pickle.load(open('../scripts/results_mnist/exp_2_R_05_config.p','rb'))
```

```python
res_m = res.mean(0)
res_std = res.std(0)
```

```python
units_list = [20,30,40,50,60,70,80,90,100]

names = ['Standard','Weighted','LC']
colors = ['blue','red','green']
plt.figure()
plt.title('Utility over test data')
M = 1.0001
i=0
for name in names: 
    std = res_std[:,i,1]
    plt.plot(units_list[:],M-res_m[:,i,1],label=name,color=colors[i])
    plt.plot(units_list[:],M-res_m[:,i,1]+std,color=colors[i],alpha=0.2)
    plt.plot(units_list[:],M-res_m[:,i,1]-std,color=colors[i],alpha=0.2)
    i+=1
plt.legend(loc=0)
plt.xlabel('Capacity of the network / number of units in hidden layer')
# plt.savefig("../scripts/results_mnist/util_exp_2_R05_M.pdf",bbox_inches='tight',rasterized=True)
plt.show()
```
