```python
import sys
sys.path.insert(0, '../code/')
sys.path.insert(0, '../data/')
import lcbnn_regression as lcbnn
import synthetic
from utils import plot_confusion_matrix
import tensorflow as tf
import numpy as np
import random
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

import matplotlib.pyplot as plt
import matplotlib.cm as cm
%matplotlib inline

# Keras
import keras
from keras import backend as K
from keras.models import Sequential, Model
from keras.metrics import binary_accuracy
from keras.layers import Input
from keras.layers.core import Dense, Dropout, Activation, Layer, Lambda
from keras.regularizers import l2
from keras.optimizers import RMSprop, Adadelta, adam

# Settings
plt.rcParams['figure.figsize'] = (10, 10) # Width and height
seed = 3 #1
np.random.seed(seed=seed)
random.seed(seed)
batch_size = 50
epochs = 200#50
```

## Import data

```python
noise = 0.2
N_train = 500
N_val = 100
N_test = 500
bound = 0.5 #0.8

x_train = np.linspace(0,4*np.pi,N_train).reshape((N_train,1))
y_train = np.sin(x_train) + np.random.multivariate_normal([0],[[noise]],(N_train))

x_val = np.linspace(0,4*np.pi,N_val).reshape((N_val,1))
y_val = np.sin(x_val) #+ np.random.multivariate_normal(0,noise,(N_train,))

x_test = np.linspace(0,4*np.pi,N_test).reshape((N_test,1))
y_test = np.sin(x_test) #+ np.random.multivariate_normal(0,noise,(N_train,))


x_train = (x_train - x_train.mean())/x_train.std()
x_test = (x_test - x_test.mean())/x_test.std()
x_val = (x_val - x_val.mean())/x_val.std()

# y_train = (y_train - y_train.mean())/y_train.std()
# y_test = (y_test - y_test.mean())/y_test.std()
# y_val = (y_val - y_val.mean())/y_val.std()

```

```python
ax1 = plt.figure()
plt.plot(x_train,y_train, label='Train')

plt.plot([np.min(x_train),np.max(x_train)],[bound,bound],'tab:red',label='T_crit')
plt.ylabel('Temperature')
plt.xlabel('Input')
plt.legend()

plt.show()
```

## Set up utility

```python
num_classes = 2
alpha = 100 #10
M = 10.00001
loss_mat = np.array([[0.,10.],[1.,0.0]]).astype(np.float32) #np.array([[0.,10.],[1.,0.0]]).astype(np.float32)
print("         TRUTH\n" )
print('P  ',loss_mat[0])
print('R  ',loss_mat[1])

# Set up weight for weighted cross entropy
class_weight = {0 : 1,
    1: 2}

# Set up bounds
def sigmoid(x):
    return 1/(1+np.exp(-x))

sig1_tf = lambda x: tf.cast(tf.sigmoid((x - bound)*alpha),dtype=tf.float32)
sig_not_1_tf = lambda x: tf.cast(1. - tf.sigmoid((x - bound)*alpha),dtype=tf.float32)
reg_class_mask_list_tf = [sig_not_1_tf,sig1_tf]

sig1 = lambda x:sigmoid((x - bound)*alpha)
sig_not_1 = lambda x: 1. - sigmoid((x - bound)*alpha)
reg_class_mask_list = [sig_not_1,sig1]

a = 1/(2*noise**2)

gammas = [a,a,a]
gamma_schedule = [100,150,200]
```

## General network structure

```python
Early_Stop_acc = keras.callbacks.EarlyStopping(monitor='val_acc', min_delta=0, patience=10, verbose=1, mode='auto')
# units
units_list = [500,500]#,50] #[1000,1000,1000]

dropout=0.1
# Regularise
tau = 5 #1.0
lengthscale = 0.2 #.01
reg = lengthscale**2 * (1 - dropout) / (2. * N_train * tau)

def network():
    model = Sequential()
    model.add(Dense(units_list[0], activation='relu', input_shape=x_train.shape[1:],W_regularizer=l2(reg)))
    model.add(Lambda(lambda x: K.dropout(x,level=dropout)))
    for units in units_list[1:]:
        model.add(Dense(units, activation='relu',W_regularizer=l2(reg)))
        model.add(Lambda(lambda x: K.dropout(x,level=dropout)))
    model.add(Dense(y_train.shape[1], activation='linear',W_regularizer=l2(reg)))
    return model
```

### Standard

```python
weights_file_std = '../models/reg/reg_weights_std.h5'
model_checkpoint =  keras.callbacks.ModelCheckpoint(weights_file_std, monitor='val_loss', save_best_only=True,
                                   save_weights_only=True, mode='auto',verbose=0)



std_model = network()
std_model.compile(loss='mean_squared_error',
                  optimizer='adam',
                  metrics=['accuracy'])

history_std = std_model.fit(x_train, y_train,
                  batch_size=batch_size,
                  epochs=epochs,
                  verbose=0,
                  validation_data=[x_val,y_val],
                  callbacks=[model_checkpoint])

std_model.load_weights(weights_file_std)
```

### Loss-calibrated

```python
weights_file_lc = '../models/reg/reg_weights_lc.h5'
model_checkpoint =  keras.callbacks.ModelCheckpoint(weights_file_lc, monitor='val_loss', save_best_only=True,
                                   save_weights_only=True, mode='auto',verbose=0)

M_tf = tf.constant(M, dtype=tf.float32)
basic_model = network()
# Initilise loss
loss = lcbnn.loss_K(loss_mat,reg_class_mask_list_tf)

H_x = Input(shape=(num_classes,),name='H_x')
y_true = Input(shape=(y_train.shape[1],),name='y_true')
gamma = Input(shape=(1,),name='gamma')
x = Input(shape=(x_train.shape[1:]),name='x')

y_pred = basic_model(x)

lc_model = Model([x,H_x,gamma],y_pred)
lc_model.compile(loss = lcbnn.cal_loss(loss,M,H_x,gamma),optimizer='adam')

#Initiailisations:
y_pred_samples = np.expand_dims(y_train,0)
y_pred_samples_val = np.expand_dims(y_val,0)
H_x = lcbnn.optimal_h(y_pred_samples,loss_mat,reg_class_mask_list) #np.random.randint(0,2,(np.shape(y_train)))
H_x_val = lcbnn.optimal_h(y_pred_samples_val,loss_mat,reg_class_mask_list)

history_lc = []
g = 0
gamma_x = np.zeros_like(y_train)
gamma_x[:] = gammas[g]
gamma_x_val = np.zeros_like(y_val)
gamma_x_val[:] = gammas[g]
for epoch in range(epochs):
    h_lc = lc_model.fit([x_train,H_x,gamma_x],[y_train],
                 batch_size=batch_size,
                 nb_epoch=1,
                 verbose=0,
                 validation_data=[[x_val,H_x_val,gamma_x_val],y_val],
                 callbacks=[model_checkpoint])
    T = 10
    yt_hat = np.array([lc_model.predict([x_train,H_x,gamma_x]) for _ in range(T)])
    H_x = lcbnn.optimal_h(yt_hat,loss_mat,reg_class_mask_list)
    history_lc.append(h_lc)
    if epoch % 20 ==1:
        print('Epoch: ',epoch)
    if epoch > gamma_schedule[g]:
        g += 1
        if g == len(gammas):
            g = 1 - len(gammas)
        gamma_x[:] = gammas[g]
        gamma_x_val[:] = gammas[g] 
    
lc_model.load_weights(weights_file_lc)
```

```python
def loss_K(loss_mat,reg_class_mask_list):
    """
    TF function to calculate loss from the loss matrix:
    Inputs:
        y_true: true values (N,D)
        y_pred: predicted values (N,D)
        loss_mat: matrix of loss values for selecting outputs (D,D)
        reg_class_mask_list : list of functions to mask the output into classifiction predictions
    """
#     loss_mat = tf.constant(loss_mat,dtype=tf.float32)
    # N = tf.shape(y_pred)[0] # TODO make this work for multiple inputs
#     _,D = tf.shape(loss_mat)
    D = len(reg_class_mask_list)
    def loss(y_true,y_pred): # Note y_pred in the nn is h_x
        y_true_class_l = []
        for d in range(D):
            y_true_class_l.append(reg_class_mask_list[d](y_true))
        y_true_class = tf.concat(y_true_class_l, 1)
        A = tf.matmul(y_pred, loss_mat) # Select rows with pred
        B = tf.matmul(A, tf.transpose(y_true_class))
        L = tf.diag_part(B)
        return L
    return loss

def cal_loss(loss,M,H_x):
    """
    Loss function for lcbnn regression - Squared error likelihood
    Inputs:
        loss: TF function loss(y_true,y_pred)
        M: Bound on loss (to convert to utility)
        H_x: Optimal label for each x
        reg_class_mask_list : list of functions to mask the output into classifiction predictions
    Outputs:
        dyn_loss: dynamic loss function for network
    """
    def dyn_loss(y_true,y_pred):
#         L = 1*K.square(y_pred - y_true) - K.mean(K.log(M-loss(y_pred,H_x)),axis=-1)
#         L = K.mean( K.square(y_pred - y_true) - K.log(M-loss(y_pred,H_x)),axis=-1)
        L = loss(y_pred,H_x)
#         L = - K.log(M-loss(y_pred,H_x))
        L = K.square(y_pred - y_true)
        return L
    return dyn_loss

loss = lcbnn.loss_K(loss_mat.astype(np.float32),reg_class_mask_list_tf)
```

```python
class derv_GPR(GPR):
    def __init__(self)
    
    @autoflow((settings.float_type, [None, None]))
    def predict_y(self, Xnew):
        """
        Compute the mean and variance of held-out data at the points Xnew
        """
        pred_f_mean, pred_f_var = self._build_predict(Xnew)
        pred_y_mean, pred_y_var = self.likelihood.predict_mean_and_var(pred_f_mean, pred_f_var)
        return tf.gradients(pred_y_mean,Xnew)
```
